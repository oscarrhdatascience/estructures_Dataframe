```python
# Now we group the flights by carrier so we can see a ranking of the Carriers by the Number Of Flights
```

```python
dfOrg3=dfOrg.rename(columns={"Unnamed: 0": "NumOfFlights"})
dfOrg3=dfOrg3.groupby(['UniqueCarrier']).count()
dfOrg3=dfOrg3[["NumOfFlights"]]
dfOrg3=dfOrg3.reset_index()
dfOrg3
```

```python
# We iterate the columns NumOfFlights and UniqueCarrier and save the items on a list
```

```python
flights = []
carriers = []
for index, row in dfOrg3.iterrows():
    flights.append(row['NumOfFlights'])
for index, row in dfOrg3.iterrows():
    carriers.append(row['UniqueCarrier'])
```

```python
# Then we use matplotlib library to make a plot so we can see better how the flights are distributed by the Carriers 
```

```python
plt.figure(figsize=(14,4.8))
plt.bar(carriers, flights)
plt.ylabel('Numero de vuelos anuales')
plt.xlabel('Aerolinias')
plt.title('Numero de vuelos total por aerolinia')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.pie(flights, labels=carriers, autopct="%0.1f %%")
plt.axis("equal")
plt.show()
```
